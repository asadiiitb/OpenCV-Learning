```python
# Histogram 
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Resources/Photos/cats.jpg')
#cv.imshow()

# Histogram plot for gray scale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray Cats', gray)

# Mask
blank = np.zeros(img.shape[:2], dtype = 'uint8')
circle = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
#cv.imshow('circle', cicle)

mask = cv.bitwise_and(gray, gray, mask = circle)
cv.imshow('Mask', mask)
# Format : cv.calcHist( image as list bcaz it calculats for images, 
#            Number of chanles : index to calculate histogram for,
#            Msak, Hist size: number of bins wrapped as list, Range of all pixel values )
gray_hist = cv.calcHist([gray], [0], None, [256], [0,256])
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()



cv.waitKey(0)
```

```python
#COlor Histogram

img = cv.imread('Resources/Photos/cats.jpg')
#cv.imshow()

# Mask
blank = np.zeros(img.shape[:2], dtype = 'uint8')
mask = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
#cv.imshow('circle', cicle)

masked = cv.bitwise_and(img, img, mask = mask)
cv.imshow('Mask', masked)

plt.figure()
plt.title('Color Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')

colors = ('b', 'g', 'r')
#Plot hitogram for each BGR channel color intencity
for i,col in enumerate(colors):
    hist = cv.calcHist([img], [i], mask, [256], [0,256])
    plt.plot(hist, color = col)
    plt.xlim([0,256])
    
plt.show()

cv.waitKey(0)
```
