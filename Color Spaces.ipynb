```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Resources\Photos\park.jpg')
cv.imshow('Park', img)

# BGR to GRAYSCALE
# Grayscale basically shows u the distribution of pixel intecities at particular pixel of image
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)

# BGR to HSV Hue Saturation Value based on humans way of color consumption
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('HSV', hsv)

#BGR to LAB or L*a*b :- 3Dimensional and covers entire range of color perception
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow('LAB', lab)

# BGR to RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow('RGB', rgb)

# For GRay to HSV :- WE need to do GRay to BGR then BGR to HSV
Gray_BGR = cv.cvtColor(gray, cv.COLOR_GRAY2BGR)

BGR_HSV = cv.cvtColor(Gray_BGR, cv.COLOR_BGR2HSV)
cv.imshow('GRAY TO HSV', BGR_HSV)

# For HSV  to gray:- WE need to do HSV to BGR then BGR to GRAY
hsv_bgr = cv.cvtColor(hsv, cv.COLOR_HSV2BGR)

bgr_gray = cv.cvtColor(hsv_bgr, cv.COLOR_BGR2GRAY)
cv.imshow('HSV to gray', bgr_gray)


plt.imshow(rgb)
plt.show()

cv.waitKey(0)
```
